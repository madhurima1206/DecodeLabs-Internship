```python
# Import required libraries
import pandas as pd
import sqlite3

# Load the cleaned Excel dataset into a DataFrame
df = pd.read_excel('C:/Users/LENOVO/Downloads/cleaned_dataset.xlsx')
df
```

```python
# Create a connection to the SQLite database
conn = sqlite3.connect('sales.db')

# Import the DataFrame into the SQLite database
df.to_sql(
    name='sales',
    con=conn,
    if_exists='replace',
    index=False
)

print("Dataset imported successfully!")
```

```python
# Define an SQL query to retrieve the first 5 records

query = "SELECT * FROM sales LIMIT 5"

result = pd.read_sql(query, conn)

print(result)
```

```python
# Data Filtering using WHERE Clause

query = """
SELECT OrderID,Product,TotalPrice
FROM sales
WHERE TotalPrice > 2500
"""

print(pd.read_sql(query, conn))
```

```python
# Count total number of records in the sales table

query = """
SELECT COUNT(*) AS Total_Orders
FROM sales
"""

print(pd.read_sql(query, conn))
```

```python
# Calculate total sales amount

query = """
SELECT SUM(TotalPrice) AS Total_Revenue
FROM sales
"""

print(pd.read_sql(query, conn))
```

```python
# Calculate average sales amount

query = """
SELECT AVG(TotalPrice) AS Average_Order_Value
FROM sales
"""

print(pd.read_sql(query, conn))
```

```python
# Calculate total quantity sold for each product

query = """
SELECT Product,
       SUM(Quantity) AS Total_Quantity
FROM sales
GROUP BY Product
"""

print(pd.read_sql(query, conn))
```

```python
# Calculate total orders for each payment method
# and display payment methods based on order frequency

query = """
SELECT PaymentMethod,
       COUNT(*) AS Total_Orders
FROM sales
GROUP BY PaymentMethod
ORDER BY Total_Orders 
"""

print(pd.read_sql(query, conn))
```
