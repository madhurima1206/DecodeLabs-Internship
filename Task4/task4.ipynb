```python
# IMport required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load Dataset
df = pd.read_excel("C:/Users/LENOVO/Downloads/cleaned_dataset.xlsx")
df
```

```python
#Order Status Distribution
plt.figure(figsize=(8,5))
sns.countplot(x='OrderStatus', data=df)
plt.title("Distribution of Order Status")
plt.xlabel("Order Status")
plt.ylabel("Number of Orders")
plt.xticks(rotation=45)
plt.show()
```

```python
#Payment Method Usage
plt.figure(figsize=(8,5))
df['PaymentMethod'].value_counts().plot(kind='bar')
plt.title("Payment Method Usage")
plt.xlabel("Payment Method")
plt.ylabel("Count")
plt.show()
```

```python
#Top Selling Products
plt.figure(figsize=(10,5))
product_sales = df.groupby('Product')['Quantity'].sum().sort_values(ascending=False)
product_sales.plot(kind='bar')
plt.title("Top Selling Products")
plt.xlabel("Product")
plt.ylabel("Total Quantity Sold")
plt.show()
```

```python
#Revenue by Product
plt.figure(figsize=(10,5))
revenue_product = df.groupby('Product')['TotalPrice'].sum().sort_values(ascending=False)
revenue_product.plot(kind='bar')
plt.title("Revenue by Product")
plt.xlabel("Product")
plt.ylabel("Revenue")
plt.show()
```

```python
#Monthly Sales Trend
df['Date'] = pd.to_datetime(df['Date'])
monthly_sales = df.groupby(df['Date'].dt.to_period('M'))['TotalPrice'].sum()

plt.figure(figsize=(12,5))
monthly_sales.plot(marker='o')
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Revenue")
plt.grid(True)
plt.show()
```

```python
#eferral Source Analysis
plt.figure(figsize=(8,5))
df['ReferralSource'].value_counts().plot(kind='bar')
plt.title("Orders by Referral Source")
plt.xlabel("Referral Source")
plt.ylabel("Orders")
plt.show()
```

```python
#Coupon Code Usage
plt.figure(figsize=(8,5))
df['CouponCode'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Coupon Code Usage")
plt.ylabel("")
plt.show()
```

```python
#Quantity vs Total Price
plt.figure(figsize=(8,5))
sns.scatterplot(x='Quantity', y='TotalPrice', data=df)
plt.title("Quantity vs Total Price")
plt.show()
```

```python
#Correlation Heatmap
plt.figure(figsize=(8,5))
sns.heatmap(
    df[['Quantity','UnitPrice','ItemsInCart','TotalPrice']].corr(),
    annot=True,
    cmap='coolwarm'
)
plt.title("Correlation Heatmap")
plt.show()
```

```python
#Revenue by Payment Method
plt.figure(figsize=(8,5))
df.groupby('PaymentMethod')['TotalPrice'].sum().sort_values(
    ascending=False
).plot(kind='bar')
plt.title("Revenue by Payment Method")
plt.xlabel("Payment Method")
plt.ylabel("Revenue")
plt.show()
```

```python
#Outliers in Total Price
plt.figure(figsize=(8,5))
sns.boxplot(y=df['TotalPrice'])
plt.title("Distribution of Total Price")
plt.ylabel("Total Price")
plt.show()
```
