```python
# Import pandas library
import pandas as pd

# Load the Excel dataset
df = pd.read_excel('C:/Users/LENOVO/Downloads/dataset.xlsx')

#Print first 5 rows of the dataset
df.head()
```

```python
#Shape of the dataset
print("Shape of the dataset :",df.shape)
print()
# Check dataset information
df.info()
```

```python
# Check for missing values
print(df.isnull().sum())
```

```python
# CouponCode column contains 309 missing values
```

```python
# Replaced missing CouponCode values with 'No Coupon'
df['CouponCode'] = df['CouponCode'].fillna('No Coupon')

#After replacing null values there are no null values
print(df.isnull().sum())
```

```python
# Check for duplicate records
print(df.duplicated().sum())
```

```python
#There are no duplicates in the dataset
```
